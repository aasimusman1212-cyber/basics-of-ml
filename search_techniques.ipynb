```python
#BFS
graph = {
'a' : ['b','c'],
'b' : ['d','e'],
'c' : ['f','g'],
'd' : [],
'e' : [],
'f' : [],
'g' : []
}
```

```python
from collections import deque #double ended queue
```

```python
def bfs(graph, start):
  visited =set()
  queue = deque([start])
  while queue:
      node = queue.popleft()
      if node not in visited:     # has this node already
        print(node , end =" ")    #leave one space
        visited.add(node)         #current node
        queue.extend(graph[node])

bfs(graph,'a')
```
