```python
#how to define the input to the machine
insurance_input_columns = [
    "age",
    "sex",
    "children",
    "smoker",
    "region",
]

insurance_target_column = "charges"

print("input columns :",insurance_input_columns)
print("target column :",insurance_target_column)
```

```python
print("duplicate rows :",insurance.duplicated().sum())
```

```python
#to remove duplicates from the data
insurance = insurance.drop_duplicates().reset_index(drop=True)
print("shape after removing duplicates :",insurance.shape)
```

```python
print("duplicate rows :",insurance.duplicated().sum())
```

```python
insurance.isnull().sum()
```

```python
#checking the skewness for the numerical columns
insurance_numerical_columns =[
    "age",
    "bmi",
    "children",
]
insurance[insurance_numerical_columns].skew()
```

```python
#filling the missing numerical values
insurance["age"] = insurance["age"].fillna(
    insurance["age"].mean()
    )
insurance["bmi"] = insurance["bmi"].fillna(
    insurance["bmi"].mean()
    )
insurance["children"] = insurance["children"].fillna(
    insurance["children"].median()
    )
```

```python
#handling the missing values in the catagorical columns
insurance["sex"] = insurance["sex"].fillna(
    insurance["sex"].mode()[0]
    )
insurance["smoker"] = insurance["smoker"].fillna(
    insurance["smoker"].mode()[0]
    )
insurance["region"] = insurance["region"].fillna(
    insurance["region"].mode()[0]
    )
```

```python
#Feature engineering
#label encoding : a preprocessing technique that converts categorical text data into unique numerical integer
#one hot encoding: it is preprocessing technique the allocate the number in the form of yes or no that is 0 or 1 by
#dividing the column respectively according to the data present in it
```

```python
#one hot encoding application
insurance_encoded = pd.get_dummies(insurance,
                                   columns=["sex","smoker","region"],dtype=int)
insurance_encoded.head()
```

```python
x = insurance_encoded.drop(
    "charges",
    axis=1)
y = insurance_encoded["charges"]
print("input shape :",x.shape)
print("input shape :",y.shape)
```

```python
#splittting the data into training and testing
```

```python
#testing and splitting the value
from sklearn.model_selection import train_test_split
x_train ,x_test, y_train, y_test = train_test_split (
  x,
  y,
  train_size=0.8,
  test_size=0.2,
random_state=42   #random is a default parameter that should be used commonly for every case
  )
print("training rows :",x_train.shape[0])
print("testing rows :",x_test.shape[0])
print("training rows :",y_train.shape[0])
print("testing rows :",y_test.shape[0])
```
